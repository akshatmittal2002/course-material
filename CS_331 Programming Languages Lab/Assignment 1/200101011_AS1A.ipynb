```python
# Write the path of file uploaded in your drive. 
with open('/content/drive/MyDrive/CS_331 Programming Languages Lab/Assignment 1A/data_file.txt', 'r') as f:
    data = f.read()
# Conversion of string list data to integer list.
raw = [int(i) for i in data.split()]
```

```python
# .................Write your BUBBLE SORT CODE here.......................
n = len(raw)
for i in range(n-1):
  for j in range(n-i-1):
    # swap if order mismatch
    if raw[j] > raw[j+1]:
      temp = raw[j]
      raw[j] = raw[j+1]
      raw[j+1] = temp
# print result to check
print(raw)
```

```python
# ........Write the sorted array into a file named "B_SORT.txt"...........
# open file in write mode
file = open('/content/drive/MyDrive/CS_331 Programming Languages Lab/Assignment 1A/B_SORT.txt','w')
# write sorted list in file
for i in raw:
  file.write(str(i)+"\n")
# close the file
file.close()
```
